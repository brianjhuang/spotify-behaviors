```python
from pyspark.sql import functions as f
from pyspark.sql import SparkSession

import os
import pandas as pd
import random
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
spark_fp = os.path.join('data', 'track_features', 'tf_mini.csv')
spark_fp2 = os.path.join('data', 'training_set', 'log_mini.csv')
```

```python
df = spark.read.load(spark_fp, format = 'csv', inferSchema = "true", header = "true")
df2 = spark.read.load(spark_fp2, format = 'csv', inferSchema = "true", header = "true")
```

```python
ids = df2.select('session_id').distinct()
```

```python
ten_ids = ids.orderBy(f.rand()).limit(100)['session_id']
#sample ten random session ids
```

```python
samp_fracs = {key:1 for key in ten_ids}
```

```python
samp_fracs
```

```python
samp_df = df2.sampleBy("session_id", fractions = samp_fracs)
```

```python
samp_df.toPandas()
```
